**1. Find all the null calue in the dataset. If there is anu in column, then fill it with the mean of that column.**

```python
car.isnull() #shows in form of boolean
```

```python
car.isnull().sum(numeric_only = True) # there are 4 null values in each column except the Cylinders column which has 6 null values
```

```python
car.info()
```

```python
# first we'll have to rename the 'MSRP' column to 'MSRP ($)' and 'Invoice' to 'Invoice ($)'.
car.rename(columns = {'MSRP' : 'MSRP ($)'}, inplace = True)
```

```python
car.rename(columns = {'Invoice' : 'Invoice ($)'}, inplace = True)
```

```python
car.head()
```

```python
car.info()
```

```python
# now we have to remove the dollar sign and spaces and change the datatype of the 'MSRP ($)' and 'Invoice ($)' columns
car["MSRP ($)"] = car["MSRP ($)"].replace(r'[\$,]', '', regex=True).astype(float)
car["Invoice ($)"] = car["Invoice ($)"].replace(r'[\$,]', '', regex=True).astype(float)
```

```python
car.head()
```

```python
car.info()
```

```python
# now to fill all null values with the mean of their respective column:
car = car.fillna(car.mean(numeric_only = True)) #only applicable to the int and float datatype columns
```
